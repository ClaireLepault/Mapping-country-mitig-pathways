```python
import pandas as pd
import numpy as np
import joblib
import matplotlib as mpl
import matplotlib.pyplot as plt
import warnings; warnings.filterwarnings(action='once')
import seaborn as sns
pd.options.display.max_columns = 500
```

### Load classification

```python
df = pd.read_csv('../3_Topic_modeling/output/database_seuil_0.02.csv')
df.head(2)
```

### Load topic file

```python
df_topics= pd.read_csv("../4_Manual_treatment_topic_table/ouput/Table_topics.csv")
df_topics.sort_values(by="T0.02",axis=0,ascending=True,inplace=True)
columns_topics = df_topics.Name.tolist()
```

### Select only topics columns in classification database

```python
df = df.iloc[:,15:54]
```

```python
df.loc['Total'] = df.sum(numeric_only=True)
```

```python
df.sort_values(by="Total",axis=1,ascending=True,inplace=True)
```

Change the original name of columns based on five terms for the final name

```python
df.columns = columns_topics
```

### Create a new dataframe counting the number of papers containing the topics pair by pair

```python
df_network = pd.DataFrame(columns = df.columns)
df_network["index_top"] = df.columns
df_network.set_index("index_top", inplace = True)
```

```python
for i in range(len(df.columns)):
    for j in range(len(df.columns)):
        df_network.loc[df.columns[i],[df.columns[j]]] = len(df.loc[(df[df.columns[i]]==1) & (df[df.columns[j]]==1) ])
df_network
```

```python
df_network = df_network.apply(pd.to_numeric)
```

### Dataframe where each cell corresponds to the number of common papers to both topics (row and column) on the total number of papers related to the topic-column)

```python
# i index of column and j index of row
for i in range(len(df.columns)):
    for j in range(len(df.columns)):
        if (i!=j):
            df_network.loc[df.columns[i],[df.columns[j]]] = df_network.loc[df_network.columns[i],[df_network.columns[j]]].values[0]/ df_network.loc[df_network.columns[j],[df_network.columns[j]]].values[0]
```

```python
for i in range(len(df.columns)):
    df_network.loc[df.columns[i],[df.columns[i]]] = 0
```

```python
df_network = df_network.round(2)
df_network 
```

### Heatmap

```python
SMALL_SIZE = 8
MEDIUM_SIZE = 10
BIGGER_SIZE = 25

plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=BIGGER_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=BIGGER_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=BIGGER_SIZE)    # fontsize of the tick labels
```

```python
sns.set(rc={'figure.figsize':(25,20)})

plt.figure()
heatmap = sns.heatmap(df_network, cmap="YlGnBu", fmt=".0f", vmin=0, vmax=0.5)
heatmap.xaxis.tick_top()
plt.xticks(rotation=90)
heatmap.set_ylabel('') 
heatmap.set_xlabel('')
plt.savefig("./output/Fig_SI7_heatmap_topics.png")
```
