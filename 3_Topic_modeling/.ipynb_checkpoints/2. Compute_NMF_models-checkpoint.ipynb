```python
import joblib
from sklearn import decomposition
```

# Non-negative Matrix Factorization

Now we decompose the TF-IDF matrix, *X*, into two matrices *W* and *H* for different combinations of parameters (*K*,$\alpha$,$l_1$) according to the objective function :

\begin{align}
\label{eq:1}
\min_{W,H \ge 0} \frac{1}{2} ||X-WH||_{Fro}^2 + \alpha \left[ l_1 \left( ||W||_1 + ||H||_1 \right) + \left( 1-l_1 \right) \left(||W||_{Fro}^2 + ||H||_{Fro}^2  \right) \right]
\end{align}

First, let's load the TF-IDF normalised document-term matrix that we stored earlier using *Joblib*:

```python
(tfidf, tfidf_feature_names) = joblib.load( "./interm/tfidf_matrix-features_names.pkl" )
print( "Loaded %d X %d document-term matrix" % (tfidf.shape[0], tfidf.shape[1]) )
```

### Create the Topic Models

We generate models for different combinations (*K*,$\alpha$,$l_1$).

We start by pre-specifying an initial range of "sensible" values:

```python
kmin, kmax = 2, 40
```

```python
def list_pas_iter(pas, iterat):
    #round to the second decimal
    value = round(0,2)
    values = [value]
    for k in range(iterat):
        value+= pas
        values.append(round(value,2))
    return(values)
```

```python
alph_list = list_pas_iter(0.01, 31) + [0.4,0.5,0.6,0.7,0.8,0.9,1]
```

```python
l1_list = list_pas_iter(0.1,10)
```

#### Apply NMF for each of these parameter combinations:

```python
topic_models = []
# try each value of k
for k in range(kmin,kmax+1):
    for alph in alph_list:
        for l1 in l1_list:
            print("Applying NMF for k=%d, alpha =%.2f, l1 = =%.1f..." % (k, alph,l1) )
            # run NMF
            model = decomposition.NMF(init="nndsvd", n_components=k, alpha=alph, l1_ratio= l1, random_state = 1511, max_iter=2000) 
            W = model.fit_transform(tfidf)
            H = model.components_
            err = model.reconstruction_err_
            # store for later
            topic_models.append( (k,alph,l1,W,H,err) )
```

### Save the 16,731 matrix decompositions for the different combinations (k, $\alpha$, $l_1$)

```python
joblib.dump((topic_models), "./interm/topic_models.pkl") 
```
